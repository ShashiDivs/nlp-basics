## Text Preprosseing

```python
import re
import nltk
from nltk.tokenize import word_tokenize
```

```python
sample = "Hey This is first class! Visit https://nlpclass.com. <b>NLP</b> Nlp spaces... $shashi !! New       "

def preprocess(text):
    # lowercasing
    text = text.lower()

    # remove urls
    text = re.sub(r'http\S+|www.\.\S+', '', text)

    # remove html tags
    text = re.sub(r'<.*?>', '', text)

    # remove punctuations
    text = re.sub(r'[^a-z0-9\s]', '', text)

    #normalize whitespaces
    text = re.sub(r'\s+', ' ', text).strip()

    return text
```

```python
clean = preprocess(sample)
```

```python
print('original:',sample)
print('preprocessed:', clean)
```
